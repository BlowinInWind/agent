```python
from ddtrace.llmobs.decorators import embedding
from letta_client import Letta

client = Letta(base_url='http://127.0.0.1:8283')


agent_state = client.agents.create(
    name="my_first_agent",
    memory_blocks=[{
      "label": "human",
      "value": "用户名字：姜通",
      "limit": 10000
    }, {
      "label": "persona",
      'value': "你是一个有好的AI助手，说话简洁有趣"
    }],
  model='zhipuai/glm-5',
  embedding='zhipuai/glm-5',
)
# 旧版: client.sources.create(name='source')
# 新版用 archives 替代 sources
archive = client.archives.create(name='source')
```

```python
# 添加单条文本到 archive
passage = client.archives.passages.create(
    archive_id=archive.id,
    text='要存入的文本内容'
)

# 批量添加文本
client.archives.passages.create_many(
    archive_id=archive.id,
    passages=[
        {'text': '第一段文本'},
        {'text': '第二段文本'},
    ]
)
```

```python
# 挂载到 agent
client.agents.archives.attach(
    agent_id='your-agent-id',
    archive_id=archive.id
)

# 从 agent 卸载
# client.agents.archives.detach(
#     agent_id='your-agent-id',
#     archive_id=archive.id
# )

# 列出所有 archives
archives = client.archives.list()
print(archives)

# 删除 archive
# client.archives.delete(archive_id=archive.id)
```
